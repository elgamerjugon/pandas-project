```python
#Importar las librerías para panda
import pandas as pd
```

```python
#Cargar los datos a partir del archivo .csv
songs_df = pd.read_csv("Top_1_000_Songs_To_Hear_Before_You_Die.csv")
songs_df
```

```python
#Obteniendo los tipos de datos de cada columna para saber si son los más adecuados
songs_df.dtypes

#Cambiando los tipos de datos para tener mejor control sobre ellos
new_types_dict = {"THEME" : str, "TITLE" : str, "ARTIST": str, "YEAR": int, "SPOTIFY_URL": str}
songs_df = songs_df.astype(new_types_dict)
songs_df.dtypes
```

```python
#Examinando si existen columnas con valores nulos y cuáles son esas columnas
columns_missing = songs_df.isnull().sum()
columns_missing

#cambiando todos los valores nulos dentro de la columna SPOTIFY_URL y cambiándolos por una búsqueda en youtube de la canción
for col in songs_df.columns:
   songs_df.loc[songs_df[col] == 'nan', col] = "https://www.youtube.com/results?search_query=" + songs_df["TITLE"] + " " + songs_df["ARTIST"]
songs_df

```

```python
print(songs_df["SPOTIFY_URL"][15])
```

```python
#Cambiando el nombre de la columna SPOTIFY_URL a un nombre más descriptivo porque ahora ya no todos los registros llevan 
#a la aplicación de Spotify

songs_df = songs_df.rename(columns={"SPOTIFY_URL":"URL"})
songs_df.columns
```

```python
#Exportando el dataframe limpio a un archivo .csv
songs_df.to_csv("Top_1_000_Songs_To_Hear_Before_You_Die_Cleaned.csv", sep = ",")
```
